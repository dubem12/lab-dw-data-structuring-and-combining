```python
# Your code goes here

import pandas as pd
import numpy as np
data1= pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
data2= pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
data3= pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
display(data1.head(4))
display(data2.head(4))
display(data3.head(4))
display(data1.columns)
display(data2.columns)
display(data3.columns)
```

```python
data1['Customer Lifetime Value'] = data1['Customer Lifetime Value'].str.replace('%', '')
#convert to float
data1['Customer Lifetime Value'] = data1['Customer Lifetime Value'].astype(float)
data1.head(4)
```

```python
#reindex the order of columns in data2 to match data1
new_column_order2= ['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']
data2= data2[new_column_order2]
#reindex the order of columns in data2 based on data1
 #check unque values in data2
data2['Policy Type'].unique()
#replace % in customer lifetime value
data2['Customer Lifetime Value']= data2['Customer Lifetime Value'].str.replace('%','')
#convert customer lifetime value to float
data2['Customer Lifetime Value']= data2['Customer Lifetime Value'].astype(float)
```

```python
data2.head(4)
```

```python
#rename columns in data3 , before reorderin the columns
data3= data3.rename(columns={'State':'ST','Gender':'GENDER'})
display(data3.columns)
#now reorder the columns in data3 to match data1
data3= data3[new_column_order2]
display(data3.columns)
```

```python
data3.head(4)
```

```python
#concatenate the three dataframes
customer_data= pd.concat([data1,data2,data3], axis=0)
```

```python
#check the shape of the new dataframe
customer_data.shape
```

```python
#check if data1,2, and data3 are in new dataframe
data1.shape[0]+data2.shape[0]+data3.shape[0]==customer_data.shape[0]
```

```python
#check if new dataframe has all the columns
customer_data.columns
#renam the columns in customer_data
customer_data.columns=['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']
```

```python
#create a function to call unique values in each column
def unique_values(data):
    for col in data.columns:
        print(col)
        print(data[col].unique())
        print('-------------------')
unique_values(customer_data)
```

```python
#reindex the columns in customer_data using reset index
customer_data.reset_index(drop=True)
```

```python
#check for nan columns
display(customer_data.isna().sum())
```

```python
#drop rows with only nan values, (add all)
customer_data=customer_data.dropna(how='all')
```

```python
#fill in missing values

#categorical values will be filled with the mode
customer_data['education'].value_counts()
#change values Bachelors to Bachelor
customer_data['education']=customer_data['education'].replace('Bachelors','Bachelor')
#check education values
customer_data['education'].value_counts()
```

```python
#check for nan values in gender
display(customer_data['gender'].unique())
#replace nan values with mode of gender
mode_gender= customer_data['gender'].mode()[0]
#replace nan values with mode
customer_data['gender'].fillna(mode_gender, inplace=True)
```

```python
#rename femal, and female as F , and male as m
customer_data['gender']=customer_data['gender'].replace({'female':'F', 'Femal':'F', 'Male':'M'})
display(customer_data['gender'].unique())
```

```python
#check for any nan value left
display(customer_data.isna().sum())
```

```python
# replace nan customer_lifetime_value with mean
customer_data['customer_lifetime_value'].fillna(customer_data['customer_lifetime_value'].mean(), inplace=True)
```

```python
#check for any nan value left
display(customer_data.isna().sum())
```
